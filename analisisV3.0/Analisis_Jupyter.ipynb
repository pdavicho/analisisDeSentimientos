```python
import pandas as pd
```

# Lasso

```python
dfLasso = pd.read_csv("lassoTweet.csv")
dfLasso.head()
```

```python
from datetime import datetime
def fechaFormat(date):
    newFecha = datetime.strftime(datetime.strptime(date,'%a %b %d %H:%M:%S +0000 %Y'), '%b-%d')
    return newFecha
```

```python
dfLasso['FechaFormato'] = [fechaFormat(date) for date in dfLasso['created at']]
dfLasso.head()
```

```python
#dfLasso["Persona"] = "Lasso"
#dfLasso.head()
dfLasso.to_csv("TweetLasso.csv")
```

```python
from plotnine import ggplot, aes, geom_line, geom_histogram
```

```python
(ggplot(dfLasso, aes(x=dfLasso['FechaFormato'], fill=dfLasso['Persona'])) + geom_histogram(position='identity',bins=20))
```

## Uniendo Frames

```python
frames = [dfLasso, dfYaku]
df = pd.concat(frames)
df.count()
```

```python
(ggplot(df, aes(x=df['FechaFormato'], fill=df['Persona'])) + geom_histogram(position='identity',bins=20))
```
